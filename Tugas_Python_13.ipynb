```
import numpy as ocan
import matplotlib.pyplot as grafik
```

```
def plmod(d, fc, area, pem, pen):

  if fc in range(150,1501):
    c1 = 69.55
    c2 = 26.16
  elif fc in range(1501,2001):
    c1 = 46.3
    c2 = 33.9

  if (area == "Urban"): 
    Cm = 0
  elif (area == "Suburban"): 
    Cm = -2*((ocan.log10((fc/28)**2))-5.4)
  elif (area == "Open"): 
    Cm = -4.78*(ocan.log10(fc))+18.33*(ocan.log10(fc))-40.94

  apen = (1.1*ocan.log10(fc) - 0.7)*pen - (1.56*ocan.log10(fc) - 1.8)
  losspath = c1 + (c2*ocan.log10(fc)) - (13.83*ocan.log10(pem)) - apen + ((44.9-6.55*ocan.log10(pem))*ocan.log(d)) - Cm
  return losspath
```

```
t_tower = 80      
t_user = 1.5      
dmax = 700          
dmin = 200          
n = 200             
loc1 = "Urban"
loc2 = "Suburban"
loc3 = "Open"
f1 = 1800           
f2 = 900            
f3 = 700            

dran=ocan.random.randint(dmin,dmax,n)
```

```
lpUrban = plmod(dran,f1,loc1,t_tower,t_user)
lpSuburban = plmod(dran,f1,loc2,t_tower,t_user)
lpOpen = plmod(dran,f1,loc3,t_tower,t_user)

lpUrban2 = plmod(dran,f2,loc1,t_tower,t_user)
lpUrban3 = plmod(dran,f3,loc1,t_tower,t_user)
```

```
sortingdran=dran[ocan.argsort(dran)]

sortingUrban=lpUrban[ocan.argsort(dran)]
sortingSuburban=lpSuburban[ocan.argsort(dran)]
sortingOpen=lpOpen[ocan.argsort(dran)]

sortingUrban2=lpUrban2[ocan.argsort(dran)]
sortingUrban3=lpUrban3[ocan.argsort(dran)]
```

```
grafik.figure()
grafik.title("Frequency 1800 MHz")
grafik.plot(sortingdran,sortingUrban, label = "Urban")
grafik.plot(sortingdran,sortingOpen, label = "Open")
grafik.plot(sortingdran,sortingSuburban, label = "Suburban")
grafik.xlabel('Jarak (m)')
grafik.ylabel('Path Loss (dB)')
grafik.legend()
grafik.show()
```

```
grafik.figure()
grafik.title("Urban Model")
grafik.plot(sortingdran,sortingUrban3, label = "700 Mhz")
grafik.plot(sortingdran,sortingUrban2, label = "900 Mhz")
grafik.plot(sortingdran,sortingUrban, label = "1800 Mhz")
grafik.xlabel('Jarak (m)')
grafik.ylabel('Path Loss (dB)')
grafik.legend()
grafik.show()
```
